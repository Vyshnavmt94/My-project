```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
X = np.arange(0,10,0.1)
#X=np.transpose(X)
X=X.reshape(100,1)
X.shape
m = 2
c = 8
n=X.shape
Y = m * X + c
#plt.plot(X,Y,'')
Yrand = -5 + np.random.random(n) * 10  # rand(n) gives values between 0 and 1, so Yrand ranges between -5 and 5
Y = Y + Yrand
#Y=Y+rand(n)
plt.plot(X, Y, '*')
h=np.ones(n)
A=np.hstack((X,h))
Q=np.linalg.pinv(A)
#R=Q.reshape(100,2)
Z=np.matmul(Q,Y)
d=np.matmul(A,Z)
plt.plot(X,d)
```
